```python
#data.txt is simply the text from pdfs parsed with the PyPDF2 module. 
#But misreads a lot of characters. Looking for better parser and will
#Replace imperfectly processed characters by corresponding utf8 char
#in order for the summarizer to perform properly 

with open('data.txt') as json_file:
    data = json.load(json_file)
```

```python
#Transforms string in list separating with line breaks.
#Possible improvement : separate lines with regexp
#according to finer criteras
def text2list(text):
    return text.split('\n')

#The heart of the problem. Creates a sliding window that determines
#whether the 3 lines are part of a paragraph,
#looking at their relative length. Returns BOOL

#Will fail to separate paragraph if the last line is almost
#the same length as others.

def paragraph_mode(textList,i,p):
    l1=len(textList[i])
    l2=len(textList[i+1])
    try:
        l3=len(textList[i+2])
    except IndexError:
        return False
    #print('l1 =',l1)
    m = statistics.median([l1, l2, l3])
    #print('m=',m)
    cond1 = abs(l1-m)/m<p
    cond2 = abs(l2-m)/m<p
    cond3 = abs(l3-m)/m<p
    return cond1 and cond2 and cond3


#Takes a list of lines and return a dictionnary of paragraphs.

def text2par(textList,p):
    count=0
    paragraph = defaultdict()
    i = 0
    while i< len(textList)-2:
        if paragraph_mode(textList,i,p)==True:
            P=''
            while paragraph_mode(textList,i,p)==True:
                        P+=textList[i]
                        #print(textList[i])
                        i +=1                        
            P+=textList[i+1]
            #print(textList[i+1])
            if i< len(textList)-2: 
                P+=textList[i+2]
                #print(textList[i+2])
            paragraph[str(count)]=P
            #print(P)
            count+=1
        else:
            i+=1
    return dict(paragraph)


#Orchestrates the whole process, from an entry of data.txt
#to a dictionnary of paragraphs 

def main(text,p):
    textList = text2list(text)
    return text2par(textList,p)
```
